```python
# Importar librerías necesarias
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# Cargar el dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Añadir la columna 'target' (maligno o benigno)
df['target'] = data.target

# Mostrar las primeras filas para ver cómo está estructurado el DataFrame
print(df.head())
print(df.describe())
df.info()
print(df.shape)
```

```python
#Ejercicio 1 - Mergear columnas
nuevo = pd.DataFrame( df, columns=['mean radius', 'mean texture'] )
otro = pd.DataFrame( df, columns=['target'] )
print(otro)
print(nuevo)
df_merged_diff_columns = pd.merge(nuevo, otro, how='outer', left_on='mean radius', right_on='target')
print(df_merged_diff_columns.head())
otro["mean radius"] = nuevo['mean radius']
print(otro)

# Fusión de ambos DataFrames por la columna 'ID' (común)
df_merged = pd.merge(nuevo, otro, how='inner', on='mean radius')
print(df_merged)
```

```python
#Ejercicio 2 - Agrupar valores:
siguiente = df_merged.groupby('target') ['mean radius'].mean()
print("Salario promedio por departamento:\n", siguiente, "\n")
siguiente = df_merged.groupby('target')['mean texture'].mean()
print("Salario promedio por departamento:\n", siguiente, "\n")
```

```python
# Ejercicio 3 - Filtrado con where()
df_where = df_merged.where(df_merged["target"] == 0)
print(df_where.head())
```

```python
# Ejercicio 4 - Filtrado con condiciones
df_where = df_merged.where(df_merged["mean radius"] > 15)
print(df_where.head())
```
